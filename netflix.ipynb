Loading the Dataset, involves importing the data into Python using Pandas.
```python
data = pd.read_csv('/content/netflix_titles  - netflix_titles .csv', encoding='latin1') # it reads the dataset from local
data
```

**Step 5:Data Cleaning:**                          
Handle missing values: Fill them with "Unknown" or drop rows/columns if necessary.
Remove duplicates or irrelevant columns.

```python
# data.isnull(): This checks for missing (null) values in the DataFrame.
#It returns a DataFrame of the same shape as data with True where the values are missing, and False where they are not.
print(data.isnull().sum())

# Fill missing values with 'Unknown' for specific columns without using inplace=True.
#fillna('Unknown'): This replaces all missing (NaN) values in the country column with the string 'Unknown'.
data['country'] = data['country'].fillna('Unknown')
data['rating'] = data['rating'].fillna('Unknown')

# Drop duplicate rows from the dataset.
#This removes duplicate rows from the DataFrame. If two or more rows have identical values across all columns, only the first one is kept.
data = data.drop_duplicates()

# Verify that missing values have been handled
#data.isnull().sum(): It gives the count of missing values for each column after the missing values were filled and duplicates removed.
print(data.isnull().sum())  # Ensure there are no missing values after cleaning
```

**Step 7:Data Visualization.**                      
Data Visualization is the graphical representation of data to help communicate insights, patterns, trends, and relationships in a dataset more effectively. Instead of looking at raw numbers or tables, visualizations like charts, graphs, and plots allow for quicker interpretation and easier understanding of complex data.

```python
# Grouping by 'rating' and counting the occurrences of each rating
rating_counts = data.groupby(['rating']).size().reset_index(name='counts')

# Display the result
print(rating_counts)
```

Explanation:                             
* The rating column contains the unique ratings.
The counts column shows how many times each rating appears in the dataset.

* Understanding Content Distribution: This operation helps to analyze how content is distributed across different ratings.

* Content Strategy: It can provide insights into whether there’s more family-friendly content (like "PG" and "TV-G") or mature content (like "TV-MA" or "R").

```python
#Creating the Pie chart based on content rating
#data.head(50):This selects the first 50 rows of the dataset (a small sample of the dataset).
data = pd.DataFrame(data.head(50))

#data['rating']:Selects the rating column from the DataFrame, which contains the content ratings (e.g., "PG", "TV-MA", "R", etc.).
#value_counts():This function counts the occurrences of each unique value in the rating column.
data['rating'].value_counts().plot(kind='pie',autopct='%1.1f%%')

plt.title("Distribution of content ratings on Netflix")
plt.show()
```

Explanation:
* Content Analysis: It helps to visually understand how content is distributed across different ratings, providing insights into the types of content available (e.g., more family-friendly or adult content).
* Quick Insights: A pie chart is a great tool to get an immediate sense of proportions without reading through detailed numbers.

```python
# It Shows Which country release most movies in each year?
# data['country']: Selects the country column from the DataFrame, which represents the countries where the movies or shows were released.

data.country.value_counts()
```

```python
#Creating a bar chart to show top 10 countries release show on Netflix
data.country.value_counts().head(10).plot(kind='bar',color='blue')
plt.title('Top 10 countries release show on Netflix')
plt.xlabel("country")
plt.ylabel('counts')
plt.show()
```

Explanation:                                                     
* Market Insights: This visualization allows you to quickly see which countries are producing the most content on Netflix.

* Content Strategy: It helps Netflix and content creators understand where the majority of their shows are coming from, guiding content investment or expansion decisions.

```python
#Top Geners listed in netflix
data.listed_in.value_counts()
```

```python
# Showing the Top 10 genres watched on Netflix
# data['listed_in']: This selects the listed_in column, which contains the genres or categories associated with each movie or TV show on Netflix.
data.listed_in.value_counts().head(10).plot(kind='barh',color="purple")

plt.title("Top 10 Geners watched on Netflix")
plt.ylabel('Geners')
plt.legend()
```

```python
# counts the Types of shows in Netflix
data.type.value_counts()
```

```python
#Creating a pie chart based on content of Types in Netflix
# data['type'].value_counts() counts how many Movies and TV Shows are in the dataset.
# .plot(kind='pie') generates a pie chart to visualize the percentage distribution of Movies vs. TV Shows.
data.type.value_counts().plot(kind='pie', autopct='%1.1f%%')

#autopct='%1.1f%%': You can add this parameter to the plot() function to show the percentage values on each slice of the pie.

plt.title("Most watched show type on Netflix")
plt.legend()
plt.show()
```

Explanation:
* Content Distribution: This pie chart gives a clear visual representation of the distribution between Movies and TV Shows, making it easy to see which type of content is more prevalent.
* Audience Preferences: It can provide insights into the type of content most users engage with, helping Netflix decide which content to produce or acquire more of.

```python
# data['release_year']: This selects the release_year column from the data DataFrame. This column contains the release years of the content.
# bins=20: This specifies the number of bins (or intervals) for the histogram.
data['release_year'].plot(kind='hist', bins=20, title='release_year')

# plt.gca(): This retrieves the current axes (the plot area) for the current figure. The gca() stands for Get Current Axes.
#.spines: The spines of a plot are the borders (top, bottom, left, right) of the plot area.
# [['top', 'right']]: This selects the top and right spines (borders) of the plot to modify.
# .set_visible(False): This hides the selected spines (top and right), making the plot look cleaner and less cluttered.
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# Most movies released in year
data.release_year.value_counts().head(20)


```

```python
#How many movies are released in each year?
# data['release_year']: Selects the release_year column, which contains the year each movie or TV show was released.
data.release_year.value_counts().head(20).plot(kind='barh',color='yellow')

plt.title('Top 10 years in which most movies are released in Netflix')
plt.xlabel("Release Year")
plt.ylabel('counts')
plt.legend()
plt.show()
```

Explanation:
* Trend Analysis: This chart shows which years had the highest number of movie releases, helping identify trends in content production.
* Visual Insights: The horizontal bar chart makes it easy to compare counts across years.
* Content Strategy: Netflix can use this analysis to evaluate past production and release trends for strategic planning.

```python
#Maximum Duration of movies/Tv shows
data.duration.value_counts().head(10)
```

```python
# data['duration']: Selects the duration column from the dataset.
# This column represents the duration of content, such as the runtime of movies (e.g., "120 min") or the number of seasons for TV shows (e.g., "2 Seasons").
# value_counts(): Counts the occurrences of each unique value in the duration column.
data.duration.value_counts().head(10).plot(kind='bar', color='gold')

plt.xlabel('Duration of Movies/ Tv shows')
plt.ylabel('count')
plt.title("Maximum Duration of Movies / Tv shows")
plt.show()
```

Explanation:
* Content Patterns: The chart helps identify common durations for movies (e.g., "90 min" or "120 min") and TV shows (e.g., "1 Season" or "2 Seasons").

* Data Distribution: It provides a quick overview of how content duration is distributed in the dataset.

* Audience Preferences: Frequent durations might align with audience preferences (e.g., short movies or limited-series TV shows).
